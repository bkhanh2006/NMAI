```python
# Tai data metro St.Peterburg
place = "Saint Petersburg, Russia"

tags_station = {
    "railway": "station",
    "station": "subway"
}

stations = ox.features_from_place(place, tags_station)

# lọc chỉ lấy Point
stations = stations[stations.geometry.type == "Point"]

print("Số ga:", len(stations))
stations.head()
```

```python
# Tai tuyen metro
tags_lines = {
    "railway": "subway"
}

lines = ox.features_from_place(place, tags_lines)

print("Số tuyến:", len(lines))
```

```python
# lấy lat lon
stations["lat"] = stations.geometry.y
stations["lon"] = stations.geometry.x

# giữ cột cần thiết
stations_clean = stations[["name", "lat", "lon"]].reset_index(drop=True)

# thêm id
stations_clean["id"] = stations_clean.index

stations_clean.head()
```
